```python
import requests
import pandas as pd
import sqlite3
import datetime
import random
import numpy as np
```

```python
# define the endpoint for openweathermap
api_key = "218451c462ad4879587ec30a1ae7dec5"
city = "Enschede"

endpoint = f"https://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric"
endpoint
```

```python
# execute the request
response = requests.get(endpoint)
```

```python
# get the data from the response
data = response.json()
```

```python
# get the weather predictions list
predictions = data['list']
```

```python
predictions[0]
```

```python
# create an empty list
temperature_list = []
dates_list = []
```

```python
# iterate through the predictions
for x in predictions:
    temp =  x['main']['temp']
    temp_date = x['dt_txt']

    temperature_list.append(temp)
    dates_list.append(temp_date)
```

```python
temperature_list[:4]
```

```python
dates_list[:4]
```

```python
# create a DataFrame from two seperate lists
df_weather = pd.DataFrame({'date': dates_list,
                           'temperature': temperature_list})
```

```python
df_weather
```

```python
# get the Adventure works data
# define the connection string
connection_string = "data/adventureworks.db"

# establish the connection
con = sqlite3.connect(connection_string)
con.text_factory = lambda b: b.decode(errors="ignore")
```

```python
# get the SalesOrderDetail data

qry_sales_order_detail = """
SELECT ModifiedDate AS date, LineTotal AS total
FROM SalesOrderDetail
"""

df_sales_order_detail = pd.read_sql_query(qry_sales_order_detail, con=con)
```

```python
len(df_sales_order_detail)
```

```python
begin_date = datetime.date(2024, 5, 31)
max_date = datetime.date(2024, 6, 5)
```

```python
begin_date + datetime.timedelta(days=4)
```

```python
# list with random dates
random_dates_list = []

day_choices = [0,1,2,3,4,5]

for x in range(0,len(df_sales_order_detail)):
    days_choice = random.choice(day_choices)
    next_date = str(begin_date + datetime.timedelta(days = days_choice))

    random_dates_list.append(next_date)
```

```python
random_dates_list
```

```python
df_sales_order_detail['date'] = random_dates_list
```

```python
df_sales_order_detail
```

```python
df_weather['date_str'] =  df_weather['date'].str[:10]
```

```python
df_weather_grouped = df_weather[['date_str', 'temperature']].groupby('date_str').mean().reset_index()
```

```python
df_weather_grouped
```

```python
df_weather_grouped['date_str'][0]
```

```python
# join the sales order detail with the weather predictions
df_weather_sales = df_sales_order_detail.merge(df_weather_grouped, how="left", left_on="date", right_on="date_str")
```
