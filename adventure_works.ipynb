```python
import pandas as pd
import sqlite3
```

```python
# define the connection string
connection_string = "data/adventureworks.db"

# establish the connection
con = sqlite3.connect(connection_string)
con.text_factory = lambda b: b.decode(errors="ignore")
```

```python
# create a query for selecting the Address
qry_sales_order_detail = """
SELECT *
FROM SalesOrderDetail
"""
```

```python
# use the query to import data from the database
df_sales_order_detail = pd.read_sql_query(qry_sales_order_detail, con)
```

```python
# country name
country = "United Kingdom"


# query the table sales_order_header_joined with address
qry_sales_order_header_address = """
SELECT s.SalesOrderID, s.OrderDate AS date, a.City AS city, a.CountryRegion AS country, s.TotalDue AS total
FROM SalesOrderHeader AS s
JOIN Address AS a
  ON s.BillToAddressID = a.AddressID
WHERE country = ?
"""

df_sales_oder_header_address = pd.read_sql_query(qry_sales_order_header_address,
                                                 con,
                                                 params=(country,))
```
